# Demo of Stochastic scenario comparison

Compare two technology scenarios stored in Pandas Dataframes (`scen1` and `scen2`). Indexes hold environmental impact categories, and columns hold stochastic estimates of these impacts, e.g., following a Monte Carlo simulation.

```python
# Typical packages
import pandas as pd
import numpy as np
import seaborn as sns
import sys

# Import module of interest
sys.path.append('/home/bill/software/lca_standard_graphs/')
import lca_standard_graphs as lsg
```

```python
# Generate fake data
impacts = ["Climate Change", "Biodiversity", "Human health", "Resources"]
offset = 1 + np.random.rand(len(impacts))
scen1 = pd.DataFrame(index=impacts, data=np.random.rand(len(impacts), 20))
scen2 = pd.DataFrame(index=impacts, data=np.random.rand(len(impacts), 20)).multiply(offset, axis=0)
```

```python
# Plot data
ax, fig = lsg.plot_stochastic_comparison(scen1, scen2, 'Technology 1', 'Technology 2')
ax.xaxis.set_label_text("Probability (%) that each technology has a lower impact")
```
